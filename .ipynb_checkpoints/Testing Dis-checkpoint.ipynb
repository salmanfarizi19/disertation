```python
import csv
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer


def scrape_url(url):
    """Scrape the webpage at the given URL and return the text data or an error."""
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            paragraphs = soup.find_all('p')
            text_data = ' '.join(p.text for p in paragraphs)
            return text_data, None  # Successfully scraped text, no error
        else:
            return None, f"Failed to retrieve the webpage, status code: {response.status_code}"
    except requests.RequestException as e:
        return None, str(e)  # Error during requests

def analyze_sentiment(text):
    """Analyze the sentiment of the provided text using VADER and return the scores."""
    sid = SentimentIntensityAnalyzer()
    return sid.polarity_scores(text)

# Specify the paths to your input and output CSV files
input_csv = 'scraped_data.csv'  # Update this to the path of your input CSV file
output_csv = 'scraped_data_with_sentiment.csv'  # Output CSV file path

# Read URLs from the input CSV
urls = []
with open(input_csv, mode='r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header if there is one
    for row in reader:
        urls.append(row[0])  # Assuming URLs are in the first column

# Process each URL, scrape the content, perform sentiment analysis, and collect the data
data = []
for url in urls:
    text, error = scrape_url(url)
    if text:  # If text was successfully scraped
        sentiment = analyze_sentiment(text)
        data.append([url, text, sentiment['neg'], sentiment['neu'], sentiment['pos'], sentiment['compound']])
    else:  # If there was an error
        data.append([url, error, None, None, None, None])

# Write the results to the output CSV file
with open(output_csv, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['URL', 'Content', 'Negative', 'Neutral', 'Positive', 'Compound'])  # Column headers
    writer.writerows(data)
```

# BERT

```python
# Initialize the BERT model for sentiment analysis
classifier = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment', truncation=True)

def scrape_url(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            text = ' '.join(p.text for p in soup.find_all('p'))
            return text, None
        else:
            return None, f"Failed to retrieve the webpage with status code {response.status_code}"
    except Exception as e:
        return None, str(e)

def analyze_sentiment_in_chunks(text, chunk_size=512):
    # Convert text to chunks
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
    scores = []
    
    for chunk in chunks:
        try:
            result = classifier(chunk)
            scores.append((result[0]['label'], result[0]['score']))
        except Exception as e:
            print("Error processing chunk: ", e)
            continue
    
    # Aggregate results from all chunks
    if scores:
        labels = [score[0] for score in scores]
        final_label = max(set(labels), key=labels.count)
        average_score = sum(score[1] for score in scores) / len(scores)
        return final_label, average_score
    return "Analysis Inconclusive", None

# Load the CSV without a header
csv_file = 'Anies.csv'
data = pd.read_csv(csv_file, header=None)

results = []
for url in data[0]:
    text, scrape_error = scrape_url(url)
    if text:
        sentiment, score = analyze_sentiment_in_chunks(text)
        results.append([url, text, sentiment, score])
    else:
        results.append([url, scrape_error, "Scraping Failed", None])

# Create a DataFrame from the results
new_data = pd.DataFrame(results, columns=['URL', 'Content', 'Sentiment', 'Score'])

# Save the updated results to a new CSV
output_csv = 'updated_sentiment_data.csv'
new_data.to_csv(output_csv, index=False)

print("Sentiment analysis completed and saved to", output_csv)

```

# Testing BERT


```python
# Initialize the BERT sentiment analysis model
classifier = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')

def scrape_content(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            return ' '.join(p.text for p in soup.find_all('p'))
        else:
            return None
    except Exception:
        return None

def analyze_sentiment(text):
    return classifier(text)

def analyze_candidates(text, candidates):
    results = {}
    for candidate in candidates:
        if candidate.lower() in text.lower():
            candidate_text = ' '.join([sentence for sentence in text.split('.') if candidate.lower() in sentence.lower()])
            if candidate_text:  # Ensure there's text to analyze
                sentiment_result = analyze_sentiment(candidate_text)
                results[candidate] = sentiment_result
    return results

# Load CSV and check if it has headers
csv_file = 'Prabowo'
try:
    data = pd.read_csv(csv_file + '.csv')
    if 'URL' not in data.columns:
        data = pd.read_csv(csv_file + '.csv', header=None, names=['URL'])  # Specify column name if no header
except Exception as e:
    print("Failed to load or process the CSV file:", str(e))
    exit()

candidates = ["Anies Baswedan", "Salman", "Prabowo Subianto", "Ganjar Pranowo"]

# Analyze each URL
analysis_results = []
for url in data['URL']:
    content = scrape_content(url)
    if content:
        sentiment_data = analyze_candidates(content, candidates)
        analysis_results.append({'URL': url, 'Sentiment Data': sentiment_data})

# Create DataFrame from results
result_df = pd.DataFrame(analysis_results)
result_df.to_csv(csv_file+'_sentiment_analysis.csv', index=False)
print("Analysis completed and saved.")

```

```python
# Load the CSV file
file_path = csv_file+'_sentiment_analysis.csv'
data = pd.read_csv(file_path)

# Display the initial structure if needed (optional)
print(data.head())

# Initialize a list to hold the parsed data
parsed_results = []

# Assume the 'Sentiment Data' column contains dictionary-like strings of sentiment results
for index, row in data.iterrows():
    url = row['URL']
    sentiment_data = eval(row['Sentiment Data'])  # Using eval to convert string to dictionary

    # Iterate through the dictionary and extract candidate and their sentiment label
    for candidate, sentiment_info in sentiment_data.items():
        # Extract the sentiment label
        label = sentiment_info[0]['label']  # Adjust based on actual structure of sentiment_info
        parsed_results.append([url, candidate, label])

# Create a DataFrame from the parsed results
result_df = pd.DataFrame(parsed_results, columns=['URL', 'Candidate', 'Label'])

# Display the final table
print(result_df)
```

```python
result_df.to_csv(csv_file+'_sentiment_analysis_parsed.csv', index=False)
```

```python
# Count the number of each label per candidate
sentiment_counts = result_df.groupby(['Candidate', 'Label']).size().unstack(fill_value=0)

# Plotting
sentiment_counts.plot(kind='bar', stacked=True, color=['red', 'gray', 'green','yellow','blue'])
plt.title('Sentiment Counts per Candidate')
plt.xlabel('Candidate')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.legend(title='Sentiment Label')
plt.tight_layout()
plt.show()
```

```python
# Load the CSV file
file_path = csv_file+'_sentiment_analysis_parsed.csv'  # Update this path as needed
df = pd.read_csv(file_path)

# Use pivot_table with aggregation to handle duplicates
pivoted_df = df.pivot_table(index='URL', columns='Candidate', values='Label', aggfunc=lambda x: ' '.join(x))

# Display the pivoted dataframe
print(pivoted_df.head())
pivoted_df.to_csv(csv_file+'_pivoted_table.csv', index=True)
```

```python
# Load the new CSV file
pivoted_file_path = csv_file+'_pivoted_table.csv'
pivoted_data = pd.read_csv(pivoted_file_path)

# Display the first few rows of the pivoted dataframe to understand its structure
pivoted_data
```

```python
# Convert sentiment ratings to numerical values for comparison
def convert_to_numeric(star_rating):
    if pd.isna(star_rating):
        return None
    return int(star_rating.split()[0])

pivoted_data['Anies Baswedan'] = pivoted_data['Anies Baswedan'].apply(convert_to_numeric)
pivoted_data['Ganjar Pranowo'] = pivoted_data['Ganjar Pranowo'].apply(convert_to_numeric)
pivoted_data['Prabowo Subianto'] = pivoted_data['Prabowo Subianto'].apply(convert_to_numeric)
pivoted_data['Salman'] = pivoted_data['Salman'].apply(convert_to_numeric)

# Calculate mean sentiment ratings for each figure
mean_sentiments = pivoted_data[['Anies Baswedan', 'Ganjar Pranowo', 'Prabowo Subianto', 'Salman']].mean()

# Plot the mean sentiment ratings
plt.figure(figsize=(10, 6))
mean_sentiments.plot(kind='bar', color=['skyblue', 'lightgreen', 'lightcoral', 'lightgray'])
plt.title('Average Sentiment Ratings Across Figures')
plt.xlabel('Figure')
plt.ylabel('Average Sentiment Rating')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
